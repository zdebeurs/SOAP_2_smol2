## Local function definitions

```python
# The following (14-Jul-2019) version of derivs operates on the rank-reduced SVD CCF,
# and calculates the second derivative from dfdv in a way that absorbs
# less of the etaloning ripple pattern. Setting use_svd=False
# reverts to using the raw CCF except in the calculation of d2fdv2.
def derivs(ccfIn,cdelt1=0.25,npc=5,use_svd=True):
    if use_svd:
        u,s,v = np.linalg.svd(ccfIn,full_matrices = False)
        ccf2d = np.dot(u[:,:npc] * s[:npc], v[:npc])
    else:
        ccf2d = ccfIn
        
    nvel = np.shape(ccf2d)[1]
    fm = np.roll(ccf2d,-1)
    f0 = np.array(ccf2d)
    fp = np.roll(ccf2d,1)
    # First derivative
    dfdv = (fm-fp) / 2 / cdelt1
    # Fix the ends
    dfdv[:,0] = dfdv[:,1]
    dfdv[:,nvel-1] = dfdv[:,nvel-2]
    # Second derivative
    dfdvm = np.roll(dfdv,-1)
    dfdvp = np.roll(dfdv,1)
    d2fdv2 = (dfdvm-dfdvp) / 2 / cdelt1
    # Fix the ends
    d2fdv2[:,0] = d2fdv2[:,1]
    d2fdv2[:,nvel-1] = d2fdv2[:,nvel-2]
    return dfdv,d2fdv2

# Construct a series of CCF corrections from a list of velocity shifts.
def vel2ccf(vseries,dfdv,d2fdv2):
    ccfVel = np.transpose(vseries*np.transpose(dfdv)+  vseries*vseries*np.transpose(d2fdv2)/2)
    return ccfVel

# Shift a time series of CCFs from the heliocentric to the barycentric frame given a list of 
# apparent solar RVs relative to solar-system barycentre.
def bary2hel(ccfBary, bsrv, cdelt1=0.25, frac=12/12):
    # Normalise (divide by np.mean of each row)
    ccfNorm = np.transpose(np.transpose(ccfBary) / np.mean(ccfBary,axis=1))

    dfdv,d2fdv2 = derivs(ccfNorm,cdelt1)
    dbsrv = (bsrv - np.mean(bsrv)) * frac
    reflex = vel2ccf(dbsrv,dfdv,d2fdv2)

    # Subtract (some fraction of) solar reflex motion 
    ccfHel = ccfNorm - reflex
    
    return np.transpose(np.transpose(ccfHel) * np.mean(ccfBary,axis=1))
```

# Combine HARPS-N solar CCF data from all years

The pickle files loaded here were constructed from daily averages of all data with differential-extinction velocity correction less than Â±4 mm/sec threshold.

```python
inpath = 'HARPS-Nsolar/'

ccfAll = np.array([])

bjd = np.array([])  # Barycentric JD (from DRS)
rvh = np.array([])  # Heliocentric RV corrected for differential extinction etc (from DRS)
bsrv = np.array([]) # Apparent solar RV relative to solar-system barycentre (from JPL Horizons)
fwhm = np.array([]) # CCF FWHM (from DRS)
cont = np.array([]) # CCF contract (from DRS)
bis = np.array([])  # CCF bisector span (from DRS)


for year in ('2015','2016','2017','2018'):
    ccfile = 'ccf_'+year
    with open(inpath+ccfile,'rb') as infile:
        ccfYear = pickle.load(infile)
        ccfAll=np.append(ccfAll,ccfYear)
        
    velfile = 'vel_'+year
    with open(inpath+velfile,'rb') as infile:
        velYear = pickle.load(infile)
        bjdYear = velYear[:,0]
        rvhYear = velYear[:,1]
        bsrvYear = velYear[:,2]
        fwhmYear = velYear[:,3]
        contYear = velYear[:,4]
        bisYear = velYear[:,5]
        bjd = np.append(bjd,bjdYear)
        rvh = np.append(rvh,rvhYear)
        bsrv = np.append(bsrv,bsrvYear)
        fwhm = np.append(fwhm,fwhmYear)
        cont = np.append(cont,contYear)
        bis = np.append(bis,bisYear)
        
nrows = np.int(size(ccfAll)/shape(ccfYear)[1])
ncols = shape(ccfYear)[1]
ccfAll=reshape(ccfAll,(nrows,ncols))
```

# Figure 1: Illustrate CCF and ACF in barycentric and heliocentric frame.

Time sequence of residual solar CCFs and the ACFs of the CCFs, spanning the period from 2015 July to 2018 December. 
Each row represents one day of observation. 
Days on which no data were obtained are not shown, so although time increases up the vertical axis, the time scale is not linear. The barycentric residual CCF (upper left) shows the solar reflex motion about the solar system barycentre. Dominated by Jupiter, it has an amplitude of 12 m~s$^{-1}$ and Jupiter's synodic period of 398 days. The residual CCF in the heliocentric frame (lower left) shows a secular change in line depth, deepening as the solar activity level declines over the period of observation.

```python
nspec, nvel = np.shape(ccfAll)
#SVDord = min(nspec,nvel)
ccfAsp = 36 / nspec

ccfBary = ccfAll

# Display results
fig, ax = plt.subplots(1,2,figsize=(10,10))

left,right,bottom,top = min(vel),max(vel),0,len(bjd)

# Compute column-averaged CCF and residuals - barycentric
ccfMean = mean(ccfBary,axis=0)
ccfRes = ccfBary - ccfMean

ax[0].imshow(ccfRes,aspect=ccfAsp,extent = (left,right,bottom,top),origin='lower')
ax[0].set_xlabel('Barycentric velocity in km s$^{-1}$')
ax[0].set_ylabel('Day number of observation')
ax[0].set_title('CCF($v$) - <CCF($v$)>')

# Transform CCF timeseries from barycentric to heliocentric frame
ccfHel = bary2hel(ccfBary, bsrv, frac=12/12)

# Compute column-averaged CCF and residuals - heliocentric
ccfMean = mean(ccfHel,axis=0)
ccfRes = ccfHel - ccfMean

ax[1].imshow(ccfRes,aspect=ccfAsp,extent = (left,right,bottom,top),origin='lower')
ax[1].set_xlabel('Heliocentric velocity in km s$^{-1}$')
ax[1].set_ylabel('Day number of observation')
ax[1].set_title('CCF($v$) - <CCF($v$)>')


plt.savefig('ccfRes.pdf',bbox_inches='tight')
plt.show()
plt.close()

print(len(ccfHel),bjd[0],bjd[len(ccfHel)-1])
```

The barycentric CCF shows the solar reflex motion about the solar system barycentre. Dominated by Jupiter, it has an amplitude of 12 m/sec and Jupiter's synodic period of 398 days.

The heliocentric CCF shows a secular change in EW. There is a an annual oscillation in profile width reflecting the Earth's orbital eccentricity, and a smaller six-month oscillation arising from the obliquity of the solar rotation axis to the earth's orbital plane.

```python
# Define the actual function where     A = p[0], mu = p[1], sigma = p[2], c = p[3] 
def gauss(x, p):
    return -p[0]*np.exp(-(x-p[1])**2/(2.*p[2]**2))+p[3]
    
# A simple minimization function:
def least(p, args):
    x, y = args
    return gauss(x, p) - y
```

# Testing MYPFIT Code Looping Throught A Bunch of Step Sizes

```python
# Define the actual function where     A = p[0], mu = p[1], sigma = p[2], c = p[3] 
def gauss(x, p):
    return -p[0]*np.exp(-(x-p[1])**2/(2.*p[2]**2))+p[3]
    
# A simple minimization function:
def least(p, args):
    x, y = args
    return gauss(x, p) - y
```

```python
# Store and plot a few of the best fit parameters

step_size_A_list = []
step_size_mu_list = []
step_size_sd_list = []
step_size_c_list = []

# Store all fit parameters to see how it corresponds
step_size_A_all_list = []
step_size_mu_all_list = []
step_size_sd_all_list = []
step_size_c_all_list = []

#for example_num in np.arange(0, number_of_ccfs):
parinfo_list=[]
number_iters_list = []
mpfit_diff_list = []
pfit_list = []
```

```python
i_list = np.random.uniform(1e-5,1e-7, size=3)
j_list = np.random.uniform(1e-2,1e-8, size=6)
k_list = np.random.uniform(1e-5,1e-17, size=6)
l_list = np.random.uniform(1e-4,1e-13, size=2)

for example_num in np.arange(0, number_of_ccfs):
    # Choose wich CCF example to examine
    # example_num = 5
    scipy_coeffs = coeff_list[example_num]
    numpy_diff = np.abs(rvh[example_num] - mu_list[example_num])

    x = np.linspace(-20, 20, 161)
    y = ccf_data_list[example_num]

    for i in i_list:
        for j in j_list:
            for k in k_list:
                for l in l_list:
                    parinfo = [{'fixed': False, 'step':i},
                               {'fixed': False, 'step':j},
                               {'fixed': False, 'step':k},
                               {'fixed': False, 'step':l}]
                    parinfo_list.append(parinfo)

                    pstart = [1., 0.1, 1., 1]
                    pfit, results = mpyfit.fit(least, pstart,(x, y), parinfo)
                    pfit_list.append(pfit)
                    
                    step_size_A_all = i
                    step_size_mu_all = j
                    step_size_sd_all = k
                    step_size_c_all = l
                    
                    step_size_A_all_list.append(step_size_A_all)
                    step_size_mu_all_list.append(step_size_mu_all)
                    step_size_sd_all_list.append(step_size_sd_all)
                    step_size_c_all_list.append(step_size_c_all)

                    number_iters= results["niter"]
                    number_iters_list.append(number_iters)
                    mpfit_diff = np.abs(rvh[example_num]-pfit[1])
                    mpfit_diff_list.append(mpfit_diff)
                    
    factor = 100000
    min_x= m.ceil(np.min(mpfit_diff_list)*factor)/factor
    x = np.where(np.array(mpfit_diff_list)<min_x)
    
    step_size_A = parinfo_list[x[0][0]][0]["step"]
    step_size_mu = parinfo_list[x[0][0]][1]["step"]
    step_size_sd = parinfo_list[x[0][0]][2]["step"]
    step_size_c = parinfo_list[x[0][0]][3]["step"]

    step_size_A_list.append(step_size_A)
    step_size_mu_list.append(step_size_mu)
    step_size_sd_list.append(step_size_sd)
    step_size_c_list.append(step_size_c)
```
